## 1. My lane (or freestyle) and why

**Refresh / Content Opportunity Scoring**

I chose the **Refresh / Content Opportunity Scoring** lane because it addresses a practical challenge in content management: deciding which webpages should be reviewed first. When websites contain hundreds or thousands of pages, reviewing each one manually is inefficient.

Through this project, I want to explore how search performance and engagement signals can help prioritize pages that may benefit from a content refresh. The goal is not to replace human judgment, but to provide data-driven recommendations that help content teams focus on the pages with the greatest opportunity for improvement.

## 2. The question: decision, action, cost of a wrong call

### Research Question

**How can search performance and engagement signals be used to prioritize webpages for content refresh?**

### Decision

The goal is to identify which webpages should be reviewed first for content updates.

### Unit of Analysis

One webpage (content item).

### Expected Output

A ranked list of webpages recommended for content review and refresh.

### Action

Content managers or SEO specialists can focus on the highest-priority webpages instead of manually reviewing every page.

### Cost of a Wrong Recommendation

If the recommendations are inaccurate, content teams may spend time updating pages with little improvement potential while overlooking pages that could have a greater impact on search performance. This can result in inefficient use of time and missed optimization opportunities.

### Why Data Can Help

Website performance is influenced by multiple factors, including impressions, clicks, click-through rate (CTR), search position, engagement, and content freshness. Looking at these signals together makes it easier to identify pages that deserve attention than relying on manual review alone.

## 3. Quick look at the data (2–3 real numbers)

I explored the starter dataset to better understand the available information before finalizing my lane. The code below provides a few summary statistics that help explain why this problem is worth investigating.

After running the code, I will summarize two to three key observations using the actual values from the dataset.

```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Display the first few rows
display(df.head())

# Basic information
print(f"Number of webpages: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")

# Trend distribution
print("\nTrend Direction")
display(df["trend_direction"].value_counts())

# Content age summary
print("\nContent Age")
display(df["content_age_days"].describe())
```

After a quick look at the starter dataset, I observed several key statistics that reinforce my choice of the Refresh / Content Opportunity Scoring lane. Firstly, the dataset comprises 30,000 individual webpages, highlighting the sheer volume of content that requires management and the impracticality of manual review. Secondly, analyzing the `content_age_days` reveals a mean age of approximately 261 days, with some content being as old as 518 days. This significant average age suggests that a substantial portion of the content may be outdated or losing relevance, presenting clear opportunities for refresh. Lastly, examining the `trend_direction` shows that while 14,436 pages are stable, a considerable 7,784 pages are experiencing a downward trend. These pages, along with stable ones, are prime candidates for a content refresh. These observations underscore the importance of a data-driven approach to prioritize webpages for update, directly addressing my research question of how search performance and engagement signals can be used to identify content opportunities.
