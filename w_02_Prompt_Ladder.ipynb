# Week 2: Prompt Ladder

## Objective

This exercise demonstrates how a weak prompt can be improved by changing **one prompt engineering layer at a time**. Each version addresses the biggest weakness of the previous output while keeping every other part of the prompt unchanged.

---

# Baseline

### Prompt

> Help me review my machine learning project.

### Representative Output

Claude could not perform the review and instead asked for additional information, such as the project files, GitHub repository, project stage, and the type of review requested.

---

# Version 1

### Layer Added

**Clear Goal**

### Prompt

> Review my machine learning project and identify its strengths, weaknesses, and the most important improvements I should make before submitting it.

### Representative Output

Claude understood that I wanted a strengths-and-weaknesses review, but it still requested the notebook or GitHub repository before providing any meaningful analysis.

### Notes

**What changed in the prompt?**

A clear review goal was added by asking for strengths, weaknesses, and priority improvements.

**What actually improved in the output?**

Very little improved. Claude better understood the type of review I wanted, but it still could not perform the task because it had no project information.

**What still failed?**

The response remained a request for additional information instead of an actual review.

**What I would try next?**

Provide real project context.

---

# Version 2

### Layer Added

**Real Context**

### Prompt

> Review my machine learning project and identify its strengths, weaknesses, and the most important improvements I should make before submitting it.
>
> The project is a machine learning workflow that includes data cleaning, exploratory data analysis, feature engineering, model training, and evaluation using regression metrics.

### Representative Output

Although Claude still requested the notebook, it also provided project-specific advice about common machine learning issues such as data leakage, evaluation metrics, baseline models, residual analysis, and cross-validation.

### Notes

**What changed in the prompt?**

Added context describing the type of machine learning project.

**What actually improved in the output?**

The response became much more specific. Instead of only asking for more information, Claude began giving practical machine learning recommendations related to the described project.

**What still failed?**

Claude still could not perform a true review because it did not have the notebook or source code.

**What I would try next?**

Specify the desired output format.

---

# Version 3

### Layer Added

**Specified Output Format**

### Prompt

> Review my machine learning project and identify its strengths, weaknesses, and the most important improvements I should make before submitting it.
>
> The project is a machine learning workflow that includes data cleaning, exploratory data analysis, feature engineering, model training, and evaluation.
>
> Present your review using the following sections:
>
> 1. Project Summary
> 2. Strengths
> 3. Weaknesses
> 4. Priority Improvements (ranked from highest to lowest impact)
> 5. Overall Assessment

*(The same notebook was attached for this and all remaining versions.)*

### Representative Output

Claude produced a structured review with separate sections for the project summary, strengths, weaknesses, ranked improvements, and an overall assessment.

### Notes

**What changed in the prompt?**

Added a required output structure.

**What actually improved in the output?**

The review became easier to read and more actionable. Recommendations were grouped logically and ranked by importance instead of appearing as general observations.

**What still failed?**

Some recommendations were broad and did not clearly separate critical issues from optional improvements.

**What I would try next?**

Add quality criteria to guide prioritization.

---

# Version 4

### Layer Added

**Quality Criteria**

### Prompt

> Review my machine learning project and identify its strengths, weaknesses, and the most important improvements I should make before submitting it.
>
> The project is a machine learning workflow that includes data cleaning, exploratory data analysis, feature engineering, model training, and evaluation.
>
> Present your review using the following sections:
>
> 1. Project Summary
> 2. Strengths
> 3. Weaknesses
> 4. Priority Improvements (ranked from highest to lowest impact)
> 5. Overall Assessment
>
> Prioritize recommendations based on their impact on model correctness, reliability, and reproducibility. For every weakness and recommendation, briefly explain why it matters.

### Representative Output

Claude justified every recommendation, explained its effect on correctness, reliability, or reproducibility, and ranked improvements according to their overall impact.

### Notes

**What changed in the prompt?**

Added quality criteria that specified how the project should be evaluated.

**What actually improved in the output?**

The recommendations became more analytical. Claude explained why each issue mattered and better distinguished high-impact methodological problems from lower-priority improvements.

**What still failed?**

Some recommendations still relied on general machine learning best practices without clearly indicating whether they were directly supported by evidence from the notebook.

**What I would try next?**

Require evidence-based verification for every claim.

---

# Version 5

### Layer Added

**Verification Requirement**

### Prompt

> Review my machine learning project and identify its strengths, weaknesses, and the most important improvements I should make before submitting it.
>
> The project is a machine learning workflow that includes data cleaning, exploratory data analysis, feature engineering, model training, and evaluation.
>
> Present your review using the following sections:
>
> 1. Project Summary
> 2. Strengths
> 3. Weaknesses
> 4. Priority Improvements (ranked from highest to lowest impact)
> 5. Overall Assessment
>
> Prioritize recommendations based on their impact on model correctness, reliability, and reproducibility. For every weakness and recommendation, briefly explain why it matters.
>
> Only make claims that are directly supported by evidence in the notebook. If a recommendation is based on an assumption or best practice rather than something observed, clearly label it as a suggestion instead of a confirmed issue.

### Representative Output

Claude labeled its findings as either **Observed** or **Suggestion**, clearly separating evidence from recommendations based on best practices. This made the review more transparent and trustworthy.

### Notes

**What changed in the prompt?**

Added a verification requirement that distinguished evidence-based observations from assumptions.

**What actually improved in the output?**

The review became more credible because confirmed findings were separated from optional recommendations. It was much easier to identify which comments were directly supported by the notebook.

**What still failed?**

Without the assignment rubric, Claude could not determine whether some optional suggestions were actually required for the project.

**What I would try next?**

If I continued improving the prompt, I would include the project requirements or grading rubric so the review could be evaluated against the intended objectives instead of only general best practices.
