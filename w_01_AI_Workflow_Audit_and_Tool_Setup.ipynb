## Task 1: Recurring Task Audit (10–15 tasks)

Classify each task as one of:
- **Just Me** — I do this myself, no AI involved.
- **Collaborate with AI** — I work interactively with AI, but I make the final decisions.
- **Delegate to AI with Review** — AI does the first pass, I review/verify before using it.
- **Fully Automate** — AI runs with no manual review needed.


| # | Recurring Task | Classification | Rationale |
|---|-----------------|-----------------|-----------|
| 1 | Taking lecture and meeting notes | Just Me | I retain information better by writing notes myself, although AI can help organize them afterward. |
| 2 | Writing social media captions | Collaborate with AI | AI helps generate ideas and improve wording, but I decide the final message and tone. |
| 3 | Grammar and writing improvement | Delegate to AI with Review | AI quickly corrects grammar and improves clarity, but I review everything before submitting. |
| 4 | Replying to important emails | Just Me | Professional emails represent me personally, so I prefer to write and review them myself. |
| 5 | Drafting emails from scratch | Collaborate with AI | AI helps structure the email, but I edit details to match the situation and recipient. |
| 6 | Creating study reviewers and summaries | Collaborate with AI | AI speeds up summarization while I verify the important concepts before studying. |
| 7 | Recording internship hours and accomplishments | Just Me | Time records must accurately reflect my actual work and cannot be delegated. |
| 8 | Writing Python code for assignments and projects | Collaborate with AI | AI helps explain concepts, debug code, and suggest improvements, but I implement and test everything myself. |
| 9 | Writing research papers and project reports | Collaborate with AI | AI assists with organization, editing, and readability, while I provide the analysis and conclusions. |
| 10 | Thesis implementation and debugging | Collaborate with AI | AI helps troubleshoot technical issues, but I design, test, and validate the implementation myself. |
| 11 | Thesis manuscript writing | Collaborate with AI | AI improves structure and writing quality, but I ensure the content accurately reflects my research. |
| 12 | Organizing datasets before analysis | Delegate to AI with Review | AI can recommend cleaning steps and identify inconsistencies, but I verify the final dataset before analysis. |
| 13 | Creating data visualizations | Collaborate with AI | AI suggests visualization methods and code, while I interpret the results and choose the final presentation. |
| 14 | GitHub documentation and README updates | Delegate to AI with Review | AI drafts documentation efficiently, but I verify that it accurately describes my project. |
