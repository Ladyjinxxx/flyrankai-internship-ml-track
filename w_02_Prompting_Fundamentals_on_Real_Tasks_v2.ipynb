# Week 2 – Prompting Fundamentals on Real Tasks v2

**Course:** AI Fluency

## Objective

The goal of this activity is to improve a weak prompt through six iterations by applying one prompt engineering technique at a time. After each iteration, I observed how the output changed. Finally, I compared the outputs generated by ChatGPT and Claude and created a reusable prompt template that can be adapted for future programming tasks.

# Task

### Selected Task
Write Python code to analyze the Iris dataset (`iris.csv`) using **pandas** and **matplotlib**.

This task was chosen because it represents a common machine learning programming assignment that requires data loading, exploratory data analysis, visualization, and interpretation.

# Version 1 – Baseline Prompt

## Prompt

```text
Write Python code to analyze this Iris dataset using pandas and matplotlib.
```

## Output Summary

ChatGPT generated a functional exploratory data analysis script that loaded the dataset, displayed summary statistics, checked for missing values, and created common visualizations such as histograms, scatter plots, box plots, and a correlation matrix.

## What Changed and Why

This served as the baseline prompt. The output successfully completed the requested task but remained generic. It focused only on producing working code without considering the intended audience, learning objectives, or preferred presentation style.

# Version 2 – Role Assignment

## Prompt

```text
You are a senior data scientist.

Write Python code to analyze this Iris dataset using pandas and matplotlib.
```

## Output Summary

The generated solution became more professional by adding clearer organization, better comments, additional validation steps, and improved visualizations. Claude also interpreted the dataset rather than only generating code.

## What Changed and Why

Assigning the AI a professional role encouraged it to produce responses that reflected the practices of an experienced data scientist. The output became more detailed, organized, and professionally structured compared to the baseline.

# Version 3 – Context and Motivation

## Prompt

```text
You are a senior data scientist.

I am a Data Science student completing a machine learning assignment. I need Python code to analyze the Iris dataset using pandas and matplotlib. The code should be beginner-friendly, well-commented, and suitable for a Google Colab notebook because I want to understand each step and explain it in my assignment.
```

## Output Summary

The response became easier to understand by providing beginner-friendly comments, clearer explanations, and a more educational coding style.

## What Changed and Why

Providing context about the user and the purpose of the task encouraged the model to tailor its response for learning rather than simply producing a working solution.

# Version 4 – Few-Shot Examples

## Prompt

```text
You are a senior data scientist.

I am a Data Science student completing a machine learning assignment. I need Python code to analyze the Iris dataset using pandas and matplotlib. The code should be beginner-friendly, well-commented, and suitable for a Google Colab notebook because I want to understand each step and explain it in my assignment.

Example of the style I want:

# Step 1: Load the dataset
# This step imports the dataset into a pandas DataFrame.
df = pd.read_csv("data.csv")

# Step 2: Display the first five rows
print(df.head())

Follow the same style throughout the solution.
```

## Output Summary

The model closely followed the provided example by organizing the notebook into consistent sections with explanatory comments and beginner-friendly formatting.

## What Changed and Why

Providing an example guided the model to imitate the desired style instead of choosing its own formatting. The response became more consistent and predictable.

# Version 5 – Output Structure

## Prompt

The previous prompt was used with the additional instruction to organize the response into the following sections:

1. Objective
2. Python Code
3. Explanation of Each Step
4. Expected Output
5. Key Takeaways

## Output Summary

The response became much more organized by separating the objective, code, explanations, expected outputs, and conclusions into clearly labeled sections.

## What Changed and Why

Explicitly specifying the desired output format improved readability and made the response resemble a structured assignment report.

# Version 6 – Step Decomposition

## Prompt

The previous prompt was used with the additional instruction to complete the task in the following order:

1. Load and inspect the dataset.
2. Perform data quality checks.
3. Compute descriptive statistics.
4. Analyze species distribution.
5. Create visualizations.
6. Compute the correlation matrix.
7. Summarize the key findings.

## Output Summary

The response followed a logical sequence from data loading to interpretation, ensuring that every important analysis step was completed before moving to the next.

## What Changed and Why

Breaking the task into ordered steps encouraged a more systematic workflow. The response became easier to follow and reduced the likelihood of skipping important parts of the analysis.

# Cross-Model Comparison

| Aspect | ChatGPT | Claude |
|--------|----------|---------|
| Tone | Professional and instructional | Conversational and educational |
| Accuracy | Generated correct exploratory data analysis code and followed the requested structure. | Generated equally accurate code while providing additional explanations and dataset-specific observations. |
| Structure | Closely followed the requested report format. | Followed the requested format while emphasizing teaching and interpretation. |
| Failure Points | Occasionally repeated information and produced longer responses. | Focused more on explanations than implementation details in some sections. |

## Overall Reflection

Both models successfully completed the task. ChatGPT consistently produced well-structured reports that closely matched the requested format, while Claude provided more educational explanations and practical insights. Depending on the goal, ChatGPT is better suited for structured documentation, whereas Claude is stronger for learning and conceptual understanding.

# Final Reusable Prompt Template

```text
Role:
You are a senior data scientist.

Context:
I am a Data Science student completing a programming assignment and need a beginner-friendly solution.

Task:
[Describe your task here.]

Example:
[Provide a short example of the preferred style.]

Output Structure:
1. Objective
2. Solution
3. Explanation
4. Expected Output
5. Key Takeaways

Complete the task in this order:
1. Understand the problem.
2. Prepare the data.
3. Perform the required analysis.
4. Explain each step.
5. Summarize the results.
```
