## 1. My lane as an ML task (type)

I am framing my project as a **scoring and ranking** task.

Instead of predicting a simple yes-or-no outcome, I want to assign each webpage a priority score based on its search performance and engagement signals. The score can then be used to rank webpages from highest to lowest priority for content refresh.

This approach is appropriate because content teams usually need to decide which pages to review first rather than simply classifying pages as "refresh" or "do not refresh."

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

df.head()
```

## 2. Target or proxy

The dataset does not contain a direct label indicating whether a webpage should be refreshed. Instead, I will use a proxy by combining available search and engagement signals to estimate refresh priority.

Rather than predicting a true outcome, the project will assign a priority score based on observed characteristics such as search trends, content age, click-through rate (CTR), impressions, and average search position. This score will represent how likely a webpage is to benefit from a content review.

```python
columns = [
    "trend_direction",
    "content_age_days",
    "ctr",
    "impression_tier",
    "avg_position"
]

df[columns].head()
```

## 3. Success metric

I will evaluate the quality of the ranking by checking whether webpages with higher priority scores consistently appear among the pages that show stronger signals for content refresh.

A good result is one where the ranking helps content teams focus on the most relevant webpages first. The goal is not perfect prediction, but a ranking that provides useful and practical decision support.

## 4. The unit of analysis, as a real dataframe

The unit of analysis is one webpage. Each row in the dataset represents a single webpage and contains search performance, engagement, and content-related features that describe its current condition.

These features will be used together to estimate a priority score for content refresh.

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.head(3)
```

## 5. Why ML beats a fixed rule here

A simple rule, such as refreshing every webpage older than one year or every page with declining traffic, ignores the fact that webpage performance depends on multiple factors working together.

For example, an older webpage may still perform well, while a newer webpage may already show signs of declining performance. Search volume, click-through rate, impressions, search position, and engagement can influence whether a webpage should be reviewed.

Machine learning can learn patterns from multiple features at the same time, making it better suited for prioritizing webpages than relying on a single fixed rule.

```python
df[[
    "content_age_days",
    "trend_direction",
    "search_volume",
    "ctr"
]].sample(5)
```
