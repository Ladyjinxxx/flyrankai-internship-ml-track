## Step 5: Pressure-Test Prompt

**Prompt used:**

I am building a portfolio for recruiters hiring Data Science and Machine Learning interns, or data entry level work.
Target Audience: Recruiters and hiring managers looking for interns or entry-level data professionals.
Primary Action: Contact me for internship or entry-level opportunities.
Proof Statement:
I build practical data science and machine learning solutions, communicate insights clearly, and continuously improve my skills through real-world internships, research, and community involvement.
Portfolio Sitemap:

* Home (Hero)
* Featured Projects / Case Studies
* About
   * Education
   * Skills & Technologies
   * Certifications
   * Community Involvement
* Contact

Please evaluate my sitemap from the perspective of a recruiter. Does every page support my proof statement and encourage the visitor to contact me? Identify any unnecessary sections, suggest anything that's missing, and recommend improvements that would make my portfolio more effective for an entry-level Data Science candidate.

**Claude's output: Summarized keypoints**

Here are the key points from Claude's feedback:

### Strengths

* The sitemap is clean, simple, and recruiter-friendly.
* **Home** should quickly communicate who you are, your value, and include a clear call-to-action.
* **Featured Projects** should be the main focus because they provide the strongest evidence of your skills.
* **Education** and **Skills** are important supporting sections.
* **Community Involvement** is worth keeping because it reinforces your continuous growth and engagement.

### Suggested Improvements

* **Add an "Experience" page** to highlight internships, research, and other real-world experience. This gives recruiters a dedicated place to see your professional background.
* **Include a Resume/CV download button** on the Home and Contact pages for easy access.
* **Decide where your research belongs**—either under Experience (if tied to an internship or lab) or under Projects (if it's an academic project like your thesis).

### Certifications

* Avoid listing too many certificates.
* Only include certifications that are relevant and recognized.
* Consider placing Certifications under **Skills** instead of making it a major section.
