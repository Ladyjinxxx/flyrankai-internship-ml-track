# Case Study

## The Problem

The Wadhwani Foundation runs a Faculty Development Program that trains faculty to teach entrepreneurship. After every training batch, they collect a post-training survey, but the survey responses had not yet been systematically analyzed to identify trends and recommendations for improving future training batches.

That's what I was brought in to do. I had 674 survey responses and needed to figure out what participants actually thought of the program: what was working, what wasn't, and what needed to change. Success meant handing over a report the Foundation could act on, not just a set of charts.

---

## What I Did

I carried out the data cleaning, analysis, visualization, interpretation, and report writing. My supervisor reviewed my work and provided feedback on the report, while I was responsible for implementing the analysis and developing the recommendations.

The data came as a CSV export from Google Forms: 674 responses with Likert-scale ratings on different parts of the program, open-ended questions, plus institution names, batch names, and timestamps. There were no demographic fields like years of teaching experience, so the analysis stayed at the overall level rather than by subgroup.

The real problem, and where most of my time went, was that 375 of the 674 responses had no timestamp. That's 56% of the dataset. Without it, I had no way to tell if a response came from the 2025 batch or the 2026 batch, which meant the year-over-year comparison the Foundation actually wanted wasn't possible.

Instead of dropping over half the data, I went back to the Consolidated Faculty Master List and matched each response to a batch name and a completion date. That recovered the training year for almost every response. Only three responses were truly unrecoverable because they were missing both the timestamp and batch name, so those were removed. I also excluded the six responses from the 2024 cohort from the year comparison because the sample was too small to produce reliable conclusions.

This step is the reason the rest of the analysis worked at all. If I'd excluded every response missing a timestamp, I would have lost most of the dataset and the comparison the Foundation cared about most would never have happened.

Once the data was usable, I evaluated the reliability of the survey using Cronbach's Alpha, which showed excellent internal consistency (α = 0.9691). I also used Pearson correlation to check for potentially redundant survey items, but the results did not support removing any questions.

To compare participant satisfaction between the 2025 and 2026 cohorts, I used Welch's t-test because the two groups had different sample sizes (449 responses in 2025 and 216 in 2026). For the qualitative responses, I performed keyword frequency analysis and created word clouds to identify recurring themes across hundreds of participant comments.

For visualization, I created bar charts to compare average scores across the five evaluation categories and all 23 survey items, a grouped bar chart for the year-over-year comparison, and word clouds for the qualitative findings.

I used Microsoft Excel during the early stages of data validation and Python in Google Colab for the analysis, using Pandas, NumPy, SciPy, Matplotlib, Seaborn, and WordCloud.

---

## What Came of It

The comparison between the 2025 and 2026 cohorts showed no statistically significant difference in participant satisfaction across any of the five evaluation categories. Although the average scores were slightly lower in 2026, the differences were not statistically significant, suggesting that participant satisfaction remained consistently high.

The qualitative analysis helped explain what participants valued and where improvements were needed. Participants consistently praised the hands-on venture creation activities, the facilitators, and the practical nature of the program. The most common concern was platform navigation, with participants frequently mentioning difficulty navigating the learning platform, finding course modules, and resuming unfinished activities. Other recurring suggestions included using more Philippine-based examples instead of Indian-context content, improving training schedules, and providing clearer facilitator instructions for online batches.

Based on these findings, I recommended creating a platform onboarding module, localizing more course content, developing an online orientation kit, revisiting facilitator support, and expanding the survey to capture additional aspects of participant experience.

I submitted the completed evaluation report to my supervisor before the end of my internship. At the time my internship concluded, I did not know whether the recommendations had been implemented, so I cannot claim organizational impact beyond delivering the report and its recommendations.

The biggest lesson I took from this project was that the hardest part of data analysis is not always the statistical methods. In this project, the biggest challenge was making incomplete data usable. Recovering the missing timestamps made the year-over-year comparison possible and changed how I think about real-world data analysis. I learned that good analysis depends just as much on careful data preparation as it does on choosing the right statistical techniques.

The project also had some limitations. The survey did not collect demographic information, so subgroup analyses were not possible. In addition, the 2026 training year was still ongoing when the analysis was conducted, so the comparison between years should be interpreted with appropriate caution.
